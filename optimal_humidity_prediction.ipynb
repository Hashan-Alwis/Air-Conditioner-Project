## Training and Testing Data

```python
from sklearn.model_selection import train_test_split
```

```python
y= df["optimal_relative_humidity"]
X= df.drop("optimal_relative_humidity", axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

#### scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Training the Model


```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score
from sklearn.tree import DecisionTreeRegressor
import xgboost as xgb
```

### XGB Regressor

```python
xg_reg = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)
```

```python
xg_reg.fit(X_train,y_train)
```

```python
predictions_xg_reg = xg_reg.predict( X_test)
```

```python
plt.scatter(y_test,predictions_xg_reg)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
plt.plot(y_test,y_test,'r')
```

### Evaluating the Model

```python
r2 = r2_score(y_test, predictions_xg_reg)

print("R^2 score:", r2)
```

```python
print('MAE:', mean_absolute_error(y_test, predictions_xg_reg))
print('MSE:', mean_squared_error(y_test, predictions_xg_reg))
print('RMSE:', np.sqrt(mean_squared_error(y_test, predictions_xg_reg)))
```

```python
sns.histplot((y_test-predictions_xg_reg),bins=50)
```

## Personal Customization Hypertune Parameters

### Hypertune Parameters tuning for Random Forest

```python
model = RandomForestRegressor()
```

```python
param_grid = {
    'bootstrap': [True],
    'max_depth': [80, 90, 100, 110],
    'max_features': [2, 3],
    'min_samples_leaf': [3, 4, 5],
    'min_samples_split': [8, 10, 12],
    'n_estimators': [100, 200, 300, 1000]
}
```

```python
grid_search = GridSearchCV(estimator = model, param_grid = param_grid, 
                          cv = 3, n_jobs = -1, verbose = 2)
```

```python
grid_search.fit(X_train, y_train)
```

```python
best_params = grid_search.best_params_
best_params
```

```python
best_model = grid_search.best_estimator_
```

```python
predictions_best_model = best_model.predict(X_test)
```

```python
plt.scatter(y_test,predictions_best_model)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
plt.plot(y_test,y_test,'r')
```

### Best model evaluation  

```python
r2 = r2_score(y_test, predictions_best_model)

print("R^2 score:", r2)
```

```python
print('MAE:', mean_absolute_error(y_test, predictions_best_model))
print('MSE:', mean_squared_error(y_test, predictions_best_model))
print('RMSE:', np.sqrt(mean_squared_error(y_test, predictions_best_model)))
```

```python
sns.histplot((y_test-predictions_best_model),bins=50)
```
